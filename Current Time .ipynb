# **Code 1**

```python
# Code 1

from datetime import datetime
import time
import pytz

ist = pytz.timezone("Asia/Kolkata")

while True:
    current_time = datetime.now(ist).strftime("%d-%m-%Y   %I:%M:%S  %p")
    print(f"\râœ… Current Date and Time (IST) :- ðŸ†— {current_time}", end="")
    time.sleep(1)
```
